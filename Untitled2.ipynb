```python
import asyncio
from playwright.async_api import async_playwright
import pandas as pd
from datetime import datetime
import nest_asyncio

nest_asyncio.apply()

async def scrape_books():
    url = "https://books.toscrape.com/"

    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        page = await browser.new_page()
        await page.goto(url, timeout=60000)

        await page.wait_for_selector('article.product_pod')

        books = await page.query_selector_all('article.product_pod')

        data = []
        for book in books[:10]:
            title_el = await book.query_selector('h3 > a')
            title = await title_el.get_attribute('title')
            link = await title_el.get_attribute('href')

            price_el = await book.query_selector('.price_color')
            price = await price_el.inner_text()

            availability_el = await book.query_selector('.availability')
            availability = (await availability_el.inner_text()).strip()

            data.append({
                "Date": datetime.now().strftime("%Y-%m-%d"),
                "Title": title,
                "Price": price,
                "Availability": availability,
                "Link": f"https://books.toscrape.com/catalogue/{link}"
            })

        return pd.DataFrame(data)

# تشغيل السكريبت
df_books = asyncio.get_event_loop().run_until_complete(scrape_books())

# حفظ البيانات
df_books.to_csv("books_data.csv", index=False)
df_books.head()
```

```python
import matplotlib.pyplot as plt

# رسم البيانات
plt.figure(figsize=(10, 5))
plt.bar(df_books['Title'], df_books['Price'], color='skyblue')
plt.xticks(rotation=75, ha='right', fontsize=9)
plt.title("Book Prices from books.toscrape.com")
plt.xlabel("Book Title")
plt.ylabel("Price (£)")
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# ⚠️ احفظ الرسم قبل plt.show()
plt.savefig("book_prices_plot.png")  # ← هذا يحفظ الصورة الصحيحة

# ثم اعرض الرسم على الشاشة
plt.show()

```
